## Problem # 2.2.

Data analysts often implement statistical inference methods by setting the probability of a
correct inference equal to 0.95. Let *A* denote the event that an inference for the population
about men is correct. Let *B* represent the event of a corresponding inference about women
being correct. Suppose that these are independent events.


(a) Find the probability that (i) *both* inferences are correct, (ii) *neither* inference is correct.  
(b) Construct the probability distribution for *Y* = number of correct inferences.  
(c) With what probability would each inference need to be correct in order for the probability to be 0.95 that *both* are correct?



`(a) Your answer goes here`

```python
pA = 0.95
pB = 0.95

p_both_correct = pA * pB
p_neither_correct = (1 - pA) * (1 - pB)

print("(a)(i) P(both correct) =", round(p_both_correct, 4))
print("(a)(ii) P(neither correct) =", round(p_neither_correct, 4))
```

`(b) Your answer goes here`

```python
pA = 0.95
pB = 0.95

# Independent Bernoulli trials
pY0 = (1 - pA) * (1 - pB)                 # neither correct
pY2 = pA * pB                              # both correct
pY1 = pA * (1 - pB) + (1 - pA) * pB        # exactly one correct

dist = {
    0: round(pY0, 4),
    1: round(pY1, 4),
    2: round(pY2, 4)
}

print("(b) Distribution of Y (number of correct inferences):")
for y, p in dist.items():
    print(f"P(Y={y}) = {p}")

print("Check sum =", round(pY0 + pY1 + pY2, 4))
```

## Problem # 2.17.

An instructor gives a course grade of B to students who have total score on exams and homeworks between 800 and 900, where the maximum possible is 1000. If the total scores have

```python
# Step 0: Create a clean income-only data file from Income.dat
# Output file: income_values.csv

import requests
import pandas as pd

url = "http://stat4ds.rwth-aachen.de/data/Income.dat"
tokens = requests.get(url, timeout=30).text.strip().split()

def is_float(x):
    try:
        float(x)
        return True
    except:
        return False

# If first token is non-numeric, assume header like: income education race
start = 0
if not is_float(tokens[0]):
    start = 3

# Parse triplets: income, education, race
income_vals = []
for i in range(start, len(tokens), 3):
    if i < len(tokens) and is_float(tokens[i]):
        income_vals.append(float(tokens[i]))

income_df = pd.DataFrame({"income": income_vals})
income_df.to_csv("income_values.csv", index=False)

print("Saved:", "income_values.csv")
print("Rows:", len(income_df))
print(income_df.head())
```

`(b) Your answer goes here`

```python
# (b) Fit a candidate distribution using same mean and SD as income values
# Recommended candidate for income: Lognormal (positive + right-skewed)

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import lognorm, norm

df = pd.read_csv("income_values.csv")
x_data = df["income"].dropna().values

m = x_data.mean()
s = x_data.std(ddof=1)

# Lognormal parameters from mean m and sd s
sigma2 = np.log(1 + (s**2 / m**2))
sigma = np.sqrt(sigma2)
mu = np.log(m) - 0.5 * sigma2

x = np.linspace(max(0.001, x_data.min()*0.5), x_data.max()*1.3, 600)

lognorm_pdf = lognorm.pdf(x, s=sigma, scale=np.exp(mu))
normal_pdf = norm.pdf(x, loc=m, scale=s)  # optional comparison

plt.figure(figsize=(9, 5))
sns.kdeplot(x_data, bw_adjust=1.0, linewidth=2.2, label="Empirical KDE")
plt.plot(x, lognorm_pdf, "r-", linewidth=2.2, label="Lognormal PDF (same mean & SD)")
plt.plot(x, normal_pdf, "g--", linewidth=1.8, label="Normal PDF (same mean & SD)")
plt.title("Income: Empirical Density vs Candidate Distribution")
plt.xlabel("Income (thousands of dollars)")
plt.ylabel("Density")
plt.legend()
plt.grid(alpha=0.25)
plt.show()

print(f"Sample mean = {m:.3f}")
print(f"Sample SD   = {s:.3f}")
print("Suggested distribution: Lognormal (income is positive and typically right-skewed).")
```

Consider the mammogram diagnostic example in Section 2.1.4.

(a) Show that the joint probability distribution of diagnosis and disease status is as shown in

of routine diagnostic screening of a population for a rare disease.

**Table 2.6.** Joint probability distribution for disease status and diagnosis of breast-cancer
mammogram, based on conditional probabilities in Table 2.1.

$$
\begin{array}{lccc}
\textbf{Disease Status} & \textbf{Positive (+)} & \textbf{Negative (–)} & \textbf{Total} \\
\hline

`Your answer goes here`

```python
# Part (a): Mammogram joint distribution + interpretation with Bayes quantities

# Given (from section/table assumptions):
# P(D) = 0.01, P(Dc)=0.99
# P(+|D)=0.86, P(-|D)=0.14
# P(+|Dc)=0.12, P(-|Dc)=0.88

P_D = 0.01
P_Dc = 0.99
P_pos_given_D = 0.86
P_neg_given_D = 0.14
P_pos_given_Dc = 0.12
P_neg_given_Dc = 0.88

# Joint probabilities
P_D_and_pos  = P_D * P_pos_given_D
P_D_and_neg  = P_D * P_neg_given_D
P_Dc_and_pos = P_Dc * P_pos_given_Dc
P_Dc_and_neg = P_Dc * P_neg_given_Dc

print("Joint table entries:")
print(f"P(D and +)   = {P_D_and_pos:.4f}")   # 0.0086
print(f"P(D and -)   = {P_D_and_neg:.4f}")   # 0.0014
print(f"P(Dc and +)  = {P_Dc_and_pos:.4f}")  # 0.1188
print(f"P(Dc and -)  = {P_Dc_and_neg:.4f}")  # 0.8712

# Positive-test totals and predictive values
P_pos = P_D_and_pos + P_Dc_and_pos
PPV = P_D_and_pos / P_pos               # P(D|+)
false_positive_share = P_Dc_and_pos / P_pos  # P(Dc|+)

print(f"\nP(+)         = {P_pos:.4f}")
print(f"P(D|+) (PPV) = {PPV:.4%}")
print(f"P(Dc|+)      = {false_positive_share:.4%}")
print(f"False positives are {P_Dc_and_pos / P_D_and_pos:.2f} times true positives.")
```

```python
# Part (b): HIV test example
# Sensitivity = 0.999, specificity = 0.9999, prevalence = 1/10000

sensitivity = 0.999
specificity = 0.9999
prevalence = 1/10000  # 0.0001

false_positive_rate = 1 - specificity

# Positive predictive value:
# PPV = P(HIV+ | test+) = [sensitivity*prevalence] /
#                         [sensitivity*prevalence + false_positive_rate*(1-prevalence)]
ppv = (sensitivity * prevalence) / (
    sensitivity * prevalence + false_positive_rate * (1 - prevalence)
)

print(f"Sensitivity: {sensitivity:.4f}")
print(f"Specificity: {specificity:.4f}")
print(f"Prevalence : {prevalence:.6f}")
print(f"PPV        : {ppv:.4%}")
```

## Problem # 2.27.

Conditional on *X = x*, suppose *Y* = weight *(kg)* has a *N*(3.0 + 0.40x, $8^2$) distribution. Simulate
and plot 1000 observations from this approximate bivariate normal distribution. Approximate
the marginal means and standard deviations for *X* and *Y*. Approximate and interpret the
correlation.


`Your answer goes here`

```python
# Simulate 1000 observations from:
# X ~ N(162, 7^2)
# Y | X=x ~ N(3.0 + 0.40x, 8^2)
# Then plot and compute marginal means/SDs and correlation.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
n = 1000

# 1) Simulate X
X = np.random.normal(loc=162, scale=7, size=n)

# 2) Simulate Y conditional on X
mu_Y_given_X = 3.0 + 0.40 * X
Y = np.random.normal(loc=mu_Y_given_X, scale=8, size=n)

# 3) Put into dataframe
df = pd.DataFrame({"Height_cm": X, "Weight_kg": Y})

# 4) Marginal summaries
mean_X = df["Height_cm"].mean()
sd_X = df["Height_cm"].std(ddof=1)
mean_Y = df["Weight_kg"].mean()
sd_Y = df["Weight_kg"].std(ddof=1)
corr_XY = df["Height_cm"].corr(df["Weight_kg"])

print(f"Approximate marginal mean of X (height): {mean_X:.2f}")
print(f"Approximate marginal SD of X (height):   {sd_X:.2f}")
print(f"Approximate marginal mean of Y (weight): {mean_Y:.2f}")
print(f"Approximate marginal SD of Y (weight):   {sd_Y:.2f}")
print(f"Approximate correlation Corr(X,Y):       {corr_XY:.3f}")

# 5) Plot: scatter + fitted line
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="Height_cm", y="Weight_kg", alpha=0.55, s=35)
sns.regplot(data=df, x="Height_cm", y="Weight_kg", scatter=False, color="red")
plt.title("Simulated Bivariate Data: Height vs Weight (n=1000)")
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.grid(alpha=0.3)
plt.show()

# 6) Optional: marginal histograms
fig, axes = plt.subplots(1, 2, figsize=(12,4))
sns.histplot(df["Height_cm"], kde=True, ax=axes[0], color="steelblue")
axes[0].set_title("Marginal Distribution of Height (X)")
axes[0].set_xlabel("Height (cm)")

sns.histplot(df["Weight_kg"], kde=True, ax=axes[1], color="darkorange")
axes[1].set_title("Marginal Distribution of Weight (Y)")
axes[1].set_xlabel("Weight (kg)")

plt.tight_layout()
plt.show()
```

## Bayes’ Theorem and AI Decision Reliability

An autonomous drone uses an onboard AI to detect whether an area contains a **target object** (for example, a missing person wearing a bright jacket).  
The system raises an alert when it believes a target is present. However, the model is not perfect.

- $P(\text{Target}) = 0.03$
- $P(\text{Alert} \mid \text{Target}) = 0.95$
- $P(\text{Alert} \mid \text{No Target}) = 0.10$

### Tasks

1. **Compute the probability of an alert**

   Use the law of total probability:
   $$P(\text{Alert}) = P(\text{Alert} \mid \text{Target})P(\text{Target}) + P(\text{Alert} \mid \text{No Target})P(\text{No Target})$$

2. **Compute the posterior probability**

   Apply Bayes’ theorem.

$$P(\text{Target} \mid \text{Alert}) =
\frac{P(\text{Alert} \mid \text{Target})P(\text{Target})}{P(\text{Alert})}$$



3. **Implement in Python**

   Write a short script that calculates and prints both probabilities with clear labels.

4. **Interpret your result**

   In a short paragraph (as a code comment), explain what your computed $(P(\text{Target} \mid \text{Alert})$ means about the AI's reliability when it triggers an alert.

5. **Bonus**

   Create a function that allows a user to change:
   - $P(\text{Target})$
   - $P(\text{Alert} \mid \text{Target})$
   - $P(\text{Alert} \mid \text{No Target})$

   Then return the updated posterior probability and show how the result changes when $P(\text{Target}) = 0.005$.

### Starter Code

```python
## Bayes' Theorem in AI Detection Reliability

############################# Given probabilities ##############################
P_target = 0.03
P_alert_given_target = 0.95
P_alert_given_no_target = 0.10
################################################################################

# Compute complements
P_no_target = 1 - P_target

# 1) Total probability of alert
P_alert = (P_alert_given_target * P_target) + (P_alert_given_no_target * P_no_target)

# 2) Posterior probability using Bayes' theorem
P_target_given_alert = (P_alert_given_target * P_target) / P_alert

print(f"P(Alert): {P_alert:.4f}")
print(f"P(Target | Alert): {P_target_given_alert:.4f} ({P_target_given_alert*100:.2f}%)")

# Interpretation:
# P(Target | Alert) is the probability a real target is present when an alert is triggered.
# Here, even though sensitivity is high (0.95), the base rate of true targets is low (0.03),
# and false alerts occur with probability 0.10 when no target is present.
# So a triggered alert is informative but not definitive; operationally, alerts should be
# followed by secondary verification (e.g., another sensor pass or human confirmation).

# Bonus: parameterized function returning both P(Alert) and posterior
def bayes_detection(P_target, P_alert_given_target, P_alert_given_no_target):
    P_no_target = 1 - P_target
    P_alert = (P_alert_given_target * P_target) + (P_alert_given_no_target * P_no_target)
    P_target_given_alert = (P_alert_given_target * P_target) / P_alert
    return P_alert, P_target_given_alert

# Bonus scenario: rarer target prevalence
P_alert_new, posterior_new = bayes_detection(0.005, 0.95, 0.10)

print(f"\nWith P(Target)=0.005:")
print(f"P(Alert): {P_alert_new:.4f}")
print(f"P(Target | Alert): {posterior_new:.4f} ({posterior_new*100:.2f}%)")
```
