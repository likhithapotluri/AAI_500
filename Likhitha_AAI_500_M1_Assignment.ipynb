```python
import pandas as pd

url = "http://stat4ds.rwth-aachen.de/data/Murder.dat"
df = pd.read_csv(url, sep=r"\s+", engine="python")
df.columns = [c.strip().lower() for c in df.columns]

state_col = [c for c in df.columns if c in ["state", "st", "abbr", "name"]][0]
murder_col = "murder"

df[murder_col] = pd.to_numeric(df[murder_col], errors="coerce")
df = df.dropna(subset=[murder_col])

# Split with and without D.C.
is_dc = df[state_col].astype(str).str.upper().isin(["DC", "D.C.", "DISTRICT_OF_COLUMBIA", "DISTRICT", "COLUMBIA"])
x_with = df[murder_col]
x_without = df.loc[~is_dc, murder_col]

# Helper stats
def summary_stats(x):
    q1 = x.quantile(0.25)
    q3 = x.quantile(0.75)
    return {
        "mean": x.mean(),
        "median": x.median(),
        "range": x.max() - x.min(),
        "iqr": q3 - q1
    }

s_without = summary_stats(x_without)
s_with = summary_stats(x_with)

print("Excluding D.C.:", {k: round(v, 3) for k, v in s_without.items()})
print("Including D.C.:", {k: round(v, 3) for k, v in s_with.items()})

delta_mean = abs(s_with["mean"] - s_without["mean"])
delta_median = abs(s_with["median"] - s_without["median"])
delta_range = abs(s_with["range"] - s_without["range"])
delta_iqr = abs(s_with["iqr"] - s_without["iqr"])

print("\nAbsolute changes when adding D.C.:")
print(f"Mean change   = {delta_mean:.3f}")
print(f"Median change = {delta_median:.3f}")
print(f"Range change  = {delta_range:.3f}")
print(f"IQR change    = {delta_iqr:.3f}")

if delta_mean > delta_median:
    print("=> Mean is affected more than median.")
else:
    print("=> Median is affected more than mean.")

if delta_range > delta_iqr:
    print("=> Range is affected more than IQR.")
else:
    print("=> IQR is affected more than range.")
```

```python
# (d) Side-by-side box plots of income by race + numerical comparison

import requests
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = "http://stat4ds.rwth-aachen.de/data/Income.dat"
text = requests.get(url, timeout=30).text.strip()
tokens = text.split()

if tokens[:3] == ["income", "education", "race"]:
    tokens = tokens[3:]

rows = []
for i in range(0, len(tokens), 3):
    rows.append((float(tokens[i]), float(tokens[i+1]), tokens[i+2]))

df = pd.DataFrame(rows, columns=["income", "education", "race"])

# side-by-side boxplots
order = ["B", "H", "W"]
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="race", y="income", order=order)
plt.title("Income by Race")
plt.xlabel("Race (B=Black, H=Hispanic, W=White)")
plt.ylabel("Income ($1000s)")
plt.grid(axis="y", alpha=0.3)
plt.show()

# numerical descriptive stats by race
group_stats = df.groupby("race")["income"].agg(
    n="count",
    mean="mean",
    median="median",
    std="std",
    min="min",
    q1=lambda s: s.quantile(0.25),
    q3=lambda s: s.quantile(0.75),
    max="max"
).reindex(order)

group_stats["IQR"] = group_stats["q3"] - group_stats["q1"]

print("Descriptive statistics by race:")
print(group_stats.round(2).to_string())
```

The `Houses` data file ([http://stat4ds.rwth-aachen.de/data/Houses.dat](http://stat4ds.rwth-aachen.de/data/Houses.dat)) at the book’s website lists the selling price (thousands of dollars), size (square feet), tax bill (dollars), number of bathrooms, number of bedrooms, and whether the house is new (1 = yes, 0 = no) for 100 home sales in Gainesville, Florida. Let’s analyze the selling prices.

(a) Construct a frequency distribution and a histogram. Describe the shape.  
(b) Find the percentage of observations that fall within one standard deviation of the mean. Why is this not close to 68%?  
(c) Construct a box plot, and interpret.   
(d) Use descriptive statistics to compare selling prices according to whether the house is new.




**Answer:**

`(a) Your answer goes here`

```python
# (a) Frequency distribution + histogram of selling price, and shape description

import pandas as pd
import matplotlib.pyplot as plt

url = "http://stat4ds.rwth-aachen.de/data/Houses.dat"
df = pd.read_csv(url, sep=r"\s+", engine="python")

# frequency distribution (choose class width = 100 in $1000s)
bins = [0,100,200,300,400,500,600,700,800,900]
freq = pd.cut(df["price"], bins=bins, right=False).value_counts().sort_index()
freq_pct = (freq / len(df) * 100).round(1)

freq_table = pd.DataFrame({
    "Price Interval ($1000s)": [str(i) for i in freq.index],
    "Frequency": freq.values,
    "Percent": freq_pct.values
})

print("Frequency Distribution of Selling Price:")
print(freq_table.to_string(index=False))

# histogram
plt.figure(figsize=(8,5))
plt.hist(df["price"], bins=bins, edgecolor="black")
plt.title("Histogram of House Selling Price")
plt.xlabel("Selling Price ($1000s)")
plt.ylabel("Frequency")
plt.grid(axis="y", alpha=0.3)
plt.show()

# quick shape indicator
print(f"\nSkewness: {df['price'].skew():.3f}")
if df["price"].skew() > 0:
    print("Shape: Right-skewed (long upper tail).")
elif df["price"].skew() < 0:
    print("Shape: Left-skewed.")
else:
    print("Shape: Approximately symmetric.")

```

```python
# (b) Percent within one standard deviation of the mean
# and why it may not be near 68%

import pandas as pd

url = "http://stat4ds.rwth-aachen.de/data/Houses.dat"
df = pd.read_csv(url, sep=r"\s+", engine="python")

mean_price = df["price"].mean()
sd_price = df["price"].std(ddof=1)

lower = mean_price - sd_price
upper = mean_price + sd_price

within_1sd = ((df["price"] >= lower) & (df["price"] <= upper)).sum()
pct_within_1sd = within_1sd / len(df) * 100

print(f"Mean price: {mean_price:.2f} ($1000s)")
print(f"SD price: {sd_price:.2f} ($1000s)")
print(f"Interval (mean ± 1 SD): [{lower:.2f}, {upper:.2f}]")
print(f"Count within interval: {within_1sd} out of {len(df)}")
print(f"Percentage within interval: {pct_within_1sd:.1f}%")

print("\nWhy not close to 68%?")
print("The 68% rule is for approximately bell-shaped (normal) distributions.")
print("House prices are often right-skewed with high-price outliers, so the rule may not hold closely.")
```

```python
# (c) Box plot + interpretation helpers

import pandas as pd
import matplotlib.pyplot as plt

url = "http://stat4ds.rwth-aachen.de/data/Houses.dat"
df = pd.read_csv(url, sep=r"\s+", engine="python")

x = df["price"]
q1 = x.quantile(0.25)
median = x.median()
q3 = x.quantile(0.75)
iqr = q3 - q1
lower_fence = q1 - 1.5 * iqr
upper_fence = q3 + 1.5 * iqr

outliers = x[(x < lower_fence) | (x > upper_fence)]

print("Five-number summary for price ($1000s):")
print(f"Min    : {x.min():.2f}")
print(f"Q1     : {q1:.2f}")
print(f"Median : {median:.2f}")
print(f"Q3     : {q3:.2f}")
print(f"Max    : {x.max():.2f}")
print(f"IQR    : {iqr:.2f}")
print(f"Outlier fences: [{lower_fence:.2f}, {upper_fence:.2f}]")
print(f"Number of outliers: {len(outliers)}")

plt.figure(figsize=(8,4))
plt.boxplot(x, vert=False)
plt.title("Box Plot of House Selling Price")
plt.xlabel("Selling Price ($1000s)")
plt.grid(axis="x", alpha=0.3)
plt.show()
```

`(c) Your answer goes here`

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# (c) Data Augmentation
# Identifying the underrepresented class (Gallstones = 1)
df_positive = df_gs[df_gs['Gallstones'] == 1]

# Generating 50 synthetic rows using simple random sampling with small noise (mimicking augmentation)
np.random.seed(42)
aug_data = {
    'Age': np.random.choice(df_positive['Age'], 50) + np.random.randint(-2, 3, 50),
    'BMI': np.random.choice(df_positive['BMI'], 50) + np.random.normal(0, 0.5, 50),
    'Cholesterol': np.random.choice(df_positive['Cholesterol'], 50) + np.random.normal(0, 5, 50),
    'Gallstones': [1] * 50
}

df_aug = pd.DataFrame(aug_data)
df_final = pd.concat([df_gs, df_aug], ignore_index=True)

# Re-running analysis to show impact
plt.figure(figsize=(12, 5))

# Before vs After Augmentation Count
plt.subplot(1, 2, 1)
sns.countplot(data=df_gs, x='Gallstones', hue='Gallstones', palette='viridis', legend=False)
plt.title('Class Distribution (Original)')

plt.subplot(1, 2, 2)
sns.countplot(data=df_final, x='Gallstones', hue='Gallstones', palette='viridis', legend=False)
plt.title('Class Distribution (Augmented)')

plt.tight_layout()
plt.show()

print(f"Original rows: {len(df_gs)}")
print(f"Augmented rows: {len(df_final)}")

# Reasoning for Augmentation:
# Augmentation 'smooths' variances by filling in sparse regions of the feature space for the minority class.
# This prevents the AI model from becoming biased toward the majority class (No Gallstones)
# and ensures more robust decision boundaries, much like GANs help in medical imaging or rare disease detection."
```
